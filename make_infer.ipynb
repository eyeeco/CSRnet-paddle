```python
import itertools
import gzip
import argparse

import paddle.v2 as paddle
import numpy as np
import paddle.v2 as paddle
from CSRNet import *
from reader import infer_data_reader, nomal
import pandas as pd
```

```python
class Struct():
    def __init__(self):
        pass
args = Struct()
args.model_gz_path='/share/dong/Result/CSR-pass-1-batch-0-test-20.6815909255.tar.gz'#path to params
args.infer_json_path = './baidu_test.json'
args.output_path='./res_epoch.csv'
logging.basicConfig()
logger = logging.getLogger("paddle")
logger.setLevel(logging.INFO)

paddle.init(use_gpu=True, trainer_count=1)
```

```python
class CSRInferer(object):
    def __init__(self,param_path):
        self.model = CSRNet(is_infer=True)
        logger.info("load model parameters from %s" % param_path)
        self.parameters = paddle.parameters.Parameters.from_tar(gzip.open(param_path, 'r'))
        self.inferer = paddle.inference.Inference(
            output_layer=self.model.model,
            parameters=self.parameters)
    def infer(self,data_path,out_path):
        logger.info("infer data...")
        infer_reader = paddle.batch(infer_data_reader(data_path), batch_size=1)
        res = []
        for idx,batch in enumerate(infer_reader()):
            temp =np.asarray(batch)
            output = self.inferer.infer(input=batch)
            res.append(output.sum())
            print(output.sum())
        pd.DataFrame(res).to_csv(out_path)
```

```python
csr_infer = CSRInferer(args.model_gz_path)
csr_infer.infer(args.infer_json_path,args.output_path)
```
